## Stratified Train/Validation Split

- Split the dataset into training and validation sets while preserving 
the class distribution
- Stratified splitting is essential for highly imbalanced datasets 
to ensure both sets contain positive samples.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
csv_path = "../data/raw_data/ISIC_2024_Training_GroundTruth.csv"
df = pd.read_csv(csv_path)

df['malignant'].value_counts()
```

```python
train_df, val_df = train_test_split(
    df,
    test_size = 0.2,
    stratify = df['malignant'],
    random_state = 42
)
```

```python
print("Train distribution:")
print(train_df['malignant'].value_counts())

print("\nValidation distribution:")
print(val_df['malignant'].value_counts())
```

```python
# Save the splits
train_df.to_csv("../data/train.csv", index=False)
val_df.to_csv("../data/val.csv", index=False)
```
