## Compute Class Weights

- Calculates class weights for handling severe class imbalance.
- The positive class weight will be used in the loss function to penalize
misclassification of malignant samples more heavily.

```python
import pandas as pd
```

```python
# Load train split
train_df = pd.read_csv("../data/train.csv")
train_df['malignant'].value_counts()
```

```python
pos = (train_df['malignant'] == 1).sum()
neg = (train_df['malignant'] == 0).sum()

pos_weight = neg / pos

print("Positive samples:", pos)
print("Negative samples:", neg)
print("pos_weight:", pos_weight)
```

## POS WEIGHT

pos_weight increases the penalty for misclassifying positive samples.
It helps the model pay more attention to rare but important cases.

```python
import json

config = {
    "pos_weight": float(pos_weight)
}

with open("../config/class_weight.json", "w") as f:
    json.dump(config, f)
```
