## Training Loop

This step implements the full training pipeline including forward pass,
loss computation, backpropagation, and validation.

It uses weighted loss to handle class imbalance.

```python
import sys
sys.path.append("..")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

from src.model import get_model
from src.dataloader import get_dataloaders
from src.train import train_one_epoch, validate
```

```python
flag = True
```

```python
# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Data
train_loader, val_loader = get_dataloaders(
    train_csv="../data/train.csv",
    val_csv="../data/val.csv",
    image_dir="../data/raw_data/ISIC_2024_Training_Input",
    batch_size=8
)
    
# Model
model = get_model().to(device)
```

```python
# Loss 
pos_weight = torch.tensor([1020.8057324840764]).to(device)  # Saved in config
loss_fn = nn.BCEWithLogitsLoss(pos_weight=pos_weight)

# Optimizer
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# Train 1 epoch test
train_loss = train_one_epoch(
    model, train_loader, optimizer, loss_fn, device,
    max_batches=20 if flag else None
)

val_loss = validate(
    model, val_loader, loss_fn, device,
    max_batches=10 if flag else None
)

print("Train loss:", train_loss)
print("Val loss:", val_loss)
```
