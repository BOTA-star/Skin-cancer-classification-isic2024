## Data Integrity check

- Verify that all image files exist and match the IDs in the metadata CSV file
- Detect missing images, incorrect filenames, or corrupted dataset structure before training

```python
import os
import pandas as pd
```

```python
csv_path = "../data/raw_data/ISIC_2024_Training_GroundTruth.csv"
df = pd.read_csv(csv_path)

print("Total samples:", len(df))
df.head()
```

```python
image_folder = "../data/raw_data/ISIC_2024_Training_Input"
print("Total images:", len(os.listdir(image_folder)))
```

```python
# Check for missing images
missing = []

for img_id in df["isic_id"]:
    img_path = os.path.join(image_folder, f"{img_id}.jpg")
    if not os.path.exists(img_path):
        missing.append(img_id)

print("Missing images:", len(missing))
```
