## Extract data

```python
import os
import zipfile
```

```python
zip_file_path = "../data/raw/ISIC_2024_Training_Input.zip"
extract_path = "../data/raw"

os.makedirs(extract_path, exist_ok=True)
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
```
